**Apriori Algorithm**

It is an algorithm that uses frequent itemset to generate association rule.It is based on the concept that subset of frequent item set must also be a frequent item set

Support: It is the frequency of item or combination of item A and B
Confidence: It is thr ratio of frequency of A and B bought together and frequency of A
Lift : tells us strength of the rule (confidence/support)

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
frequent_itemset = apriori(basket,min_support = 0.05,use_colnames = True)#Colnames give it names of column
rules = association_rules(frequent_itemset,metric = 'lift', min_threshold =1,num_itemsets= 5000)
```
